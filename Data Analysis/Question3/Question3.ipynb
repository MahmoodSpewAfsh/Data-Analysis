```python
import scipy
import numpy as np
```

```python
Data2 = np.loadtxt("0.2.txt")
Data5 = np.loadtxt("0.5.txt")
Data8 = np.loadtxt("0.8.txt")
```

```python
CD2 = scipy.signal.correlate(Data2, Data2)
CD5 = scipy.signal.correlate(Data5, Data5)
CD8 = scipy.signal.correlate(Data8, Data8)
CD25 = scipy.signal.correlate(Data2, Data5)
CD28 = scipy.signal.correlate(Data2, Data8)
CD58 = scipy.signal.correlate(Data5, Data8)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(CD2)
plt.title("auto correlation of 0.2.txt")
```

```python
plt.plot(CD5)
plt.title("auto correlation of 0.5.txt")
```

```python
plt.plot(CD8)
plt.title("auto correlation of 0.8.txt")
```

```python
plt.plot(CD25)
plt.title("correlation between 0.2.txt and 0.5.txt")
```

```python
plt.plot(CD28)
plt.title("auto correlation of 0.2.txt and 0.8.txt")
```

```python
plt.plot(CD58)
plt.title("auto correlation of 0.5.txt and 0.8.txt")
```
